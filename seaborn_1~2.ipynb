#### Seaborn tutorial
- seaborn 함수 3가지
- replot()
- scatterplot() (사용 kind="scatter";기본값
- lineplot() (kind = "line")


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```
tips = sns.load_dataset("tips")
tips
```

- seaborn에 load_dataset을 불러온다
- relplot함수를 통해 산점도 그래프를 만든다.
- x축과 y축에 2개의 변수간의 관계확인


```
sns.relplot(x="total_bill", y="tip", data=tips);
```

- 위의 x축과 y축의 관계에서 hue인수에 카테고리를 추가하여 색상을 바꾸어준다.

```
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

- style 인수를 통해서 모양을 바꾸어줄 수도 있다.

```
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips);
```

- 평면위에 두개의 인수를 모양과 색깔별로 추가할 수 있다.

```
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

- 인수안에 데이터 값이 다양하면 여러가지 색상으로 표현 될 수 있다.

```
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

- 색상을 지정 할 수 도 있다.

```
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

- size인수를 통해 카테고리에 데이터를 크기에 따라 다르게 표현 할 수도 있다.

```
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

- 사이즈를 지정해 줄 수도 있다.

```
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

```
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum())) #cumsum() null 값 무시 #ranodm.randn(500) 무작위 500개
```

```
g= sns.relplot(x="time",y="value", kind="line",data=df) #선형으로 된 #데이터 프레임 데이터
g.fig.autofmt_xdate() #그래프의 스타일 함수
```

```
fmri = sns.load_dataset("fmri")
fmri
```

```
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
#값이 복잡하여 하나의 x데이터에 다양한 y값이 존재 할수 있다 이런경우 95%신로구간을 표시한다.
```

```
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri) #신뢰구간을 끌 수 있음
```

```
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);#ci=none을 하면 표준편차의 분포를 나타냄
```

```
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri); #합계를 끄고 모든 데이터 표시
```

```
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri); #yes or no와 마찬가지로 2가지의 선택지가 있는 카테고리 하나를 추가
```

```
sns.relplot(x="timepoint", y="signal", hue="region", style="event", #선의 스타일로 도 카테고리를 추가하여 표시 가능
            kind="line", data=fmri);
```

```
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri); #선스타일 바꾸기
```

```
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri); #좀더 보기 쉽게
```

```
sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=fmri.query("event == 'stim'")); #데이터를 전부 표시
```

```
dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            kind="line", data=dots);
```

```
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            palette=palette,
            kind="line", data=dots); #생
```

```
from matplotlib.colors import LogNorm
palette = sns.cubehelix_palette(light=.7, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            hue_norm=LogNorm(),
            kind="line", data=dots);#생
```

```
sns.relplot(x="time", y="firing_rate",
            size="coherence", style="choice",
            kind="line", data=dots);
#생
```

```
sns.relplot(x="time", y="firing_rate",
           hue="coherence", size="choice",
           palette=palette,
           kind="line", data=dots);#생
```

```
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()#생
```

```
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=tips);#생
```

```
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri);#생
```

```
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"));
```
